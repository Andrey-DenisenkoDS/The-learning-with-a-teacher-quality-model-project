Продумывать подход к решению этой задачи вам не придётся — руководитель отдела его уже сформировал:

Нужно собрать данные по клиентам по следующим группам:

Создадим функцию, принимает на вход фрейм данных Pandas и возвращает подробное описание его характеристик. Она корректно обрабатывает возможные ошибки и предоставляет информативный вывод, даже если некоторые операции (например, describe для нечисловых столбцов) неприменимы.

````python
def analyze_dataframe(df, df_name="DataFrame"):

    display(f"Analysis of {df_name}:\n")

    try:
        display("Head:\n", df.head(), "\n")
    except Exception as e:
        display(f"Error displaying head: {e}\n")

    try:
        display("Info:\n", df.info(), "\n")
    except Exception as e:
        display(f"Error displaying info: {e}\n")

    try:
        display("Describe:\n", df.describe(include='all'), "\n")  # Include all data types
    except Exception as e:
        display(f"Error displaying describe: {e}\n")

    try:
        duplicates = df.duplicated().sum()
        print(f"Number of duplicate rows: {duplicates}\n")
    except Exception as e:
        print(f"Error checking for duplicates: {e}\n")

    try:
        null_counts = df.isna().mean()#isnull().sum()
        display("Null value counts per column:\n", null_counts, "\n")
    except Exception as e:
        display(f"Error counting null values: {e}\n")

    try:
        display("Data types per column:\n", df.dtypes, "\n")
    except Exception as e:
        display(f"Error displaying data types: {e}\n")

    try:
        display("Shape:\n", df.shape, "\n")
    except Exception as e:
        display(f"Error getting shape: {e}\n")


# Example usage with your three DataFrames:
analyze_dataframe(market_file, "market_file")
analyze_dataframe(market_money, "market_money")
analyze_dataframe(market_time, "market_time")
analyze_dataframe(money, "money")
````

Создадим функцию и очистим даные от столбцов и других нечисловых символов

````python
def object_to_float(df, columns):
    for col in columns:
        # Очищаем данные от пробелов и других нечисловых символов
        df[col] = df[col].str.replace(r'[^\d.]', '', regex=True)
        # Преобразуем в float, проблемные значения заменим на NaN
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

# Применение функции
columns_to_convert = ['выручка']  # добавьте все нужные колонки
market_money = object_to_float(market_money, columns_to_convert)

# Проверка результата
print(market_money['выручка'].dtype)  # должно показать float64
print(market_money['выручка'].head())

# Проверка на наличие NaN
print(market_money['выручка'].isnull().sum())
````

````python
def detailed_missing_data_report(*dataframes):
    
    for i, df in enumerate(dataframes):
        display(f"\nDataFrame {i+1}:")
        missing = df.isnull().sum()
        display(missing[missing > 0])  # Only show columns with missing values
        total_missing = missing.sum()
        percentage_missing = (total_missing / df.size) * 100 if df.size > 0 else 0
        display(f"Total missing values: {total_missing}, Percentage: {percentage_missing:.2f}%")


#Пример использования
detailed_missing_data_report(market_file, market_money, market_time, money)
````

````python
def histogram(df, col, target):
    plt.figure(figsize=(8,6))
    plot = sns.histplot(df, bins=20, kde=True, hue=target, x=col)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_ylabel('Количество', fontsize=14)
    
    
def categoral_unique(df, col):
    display(df[col].unique())
    
    plt.figure(figsize=(8,6))
    plot = sns.countplot(y=col, data=df)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_xlabel('Количество', fontsize=14)
    
    
def hist_with_wiskers(df, col, target):
    sns.set()
    f, axes = plt.subplots(1, 2, figsize=(16, 4))
    axes[0].set_title(f'Гистограмма для {col}', fontsize=16)
    axes[0].set_ylabel('Количество', fontsize=14)
    if target != None:
        sns.histplot(df, bins=20, kde=True, ax=axes[0], hue=target, x=col)
    else:
        sns.histplot(df, bins=20, kde=True, ax=axes[0], x=col)
    axes[1].set_title(f'График ящик с усами для {col}', fontsize=16)
    sns.boxplot(data=df, ax=axes[1], y=col)
    axes[1].set_ylabel(col, fontsize=14)
    plt.show()

def pivot_bar_plot(df, col):
    plt.figure(figsize=(8,6))
    plot = sns.barplot(x=col, data=df, y=df.index)
    plot.set_title(f'Рапределение по {col}', fontsize=16)
    plot.set_xlabel('Количество', fontsize=14)
````

Так же построим матрицы корреляции для значений целевого признака отдельно.

````python
def build_phik_corr_matrix(df, col, ncols=2):
    # Получаем уникальные значения целевого столбца
    unique_values = df[col].unique()
    n_plots = len(unique_values)
    
    # Рассчитываем необходимое количество строк
    nrows = int(np.ceil(n_plots / ncols))
    
    # Создаем фигуру
    fig, axs = plt.subplots(nrows=nrows, ncols=ncols, 
                           figsize=(20, 5*nrows))
    
    # Если только один график, преобразуем axs в список
    if n_plots == 1:
        axs = [axs]
    else:
        axs = axs.flatten()
    
    # Перебираем уникальные значения
    for i, value in enumerate(unique_values):
        # Фильтруем данные
        df_i = df[df[col] == value]
        
        # Проверяем, что данные не пустые
        if df_i.empty:
            continue
            
        # Строим heatmap
        try:
            sns.heatmap(df_i.phik_matrix(), 
                       annot=True, 
                       cmap='cividis', 
                       ax=axs[i],
                       vmin=0, 
                       vmax=1)
            axs[i].set_title(f'{col}: {value}')
        except Exception as e:
            print(f"Ошибка при построении для {value}: {e}")
    
    # Убираем пустые оси
    for j in range(i+1, len(axs)):
        fig.delaxes(axs[j])
    
    plt.tight_layout()
    plt.show()

# Пример использования:
#build_phik_corr_matrix(market_full, 'покупательская_активность')
````

````python
def build_phik_corr_matrix(df, col):
    ax_col = 0
    ax_row = 0
    
    fig, axs = plt.subplots(ncols=1, nrows=df[col].nunique(), figsize=(20,20))
    for i in df[col].unique():
        df_i = df[df[col] == i]
        sns.heatmap(df_i.phik_matrix(), annot=True, cmap='cividis', ax=axs[ax_col])
        axs[ax_col].set_title(i)
        ax_col+=1
        if ax_col > 1:
            ax_col=0
            ax_row+=1  
    fig.tight_layout()
    plt.show()
````

````python
def build_phik_corr_matrix(df, col):
    ax_col = 0
    ax_row = 0

    # Создаем подграфики для каждого уникального значения в столбце
    fig, axs = plt.subplots(ncols=1, nrows=df[col].nunique(), figsize=(20, 20))
    
    for i in df[col].unique():
        df_i = df[df[col] == i]
        
        # Проверка на наличие данных в подгруппе
        if not df_i.empty:
            # Вычисляем матрицу корреляции
            corr_matrix = df_i.phik_matrix()
            sns.heatmap(corr_matrix, annot=True, cmap='cividis', ax=axs[ax_col])
            axs[ax_col].set_title(f'Correlation Matrix for {i}')
        else:
            axs[ax_col].set_title(f'No data for {i}')
        
        ax_col += 1

    fig.tight_layout()
    plt.show()

# Пример использования функции
# build_phik_corr_matrix(merged_data, 'purchase_activity')
````

````python
ohe_columns = ['разрешить_сообщать', 'популярная_категория', 'тип_сервиса']
ord_columns = ['акционные_покупки_категория']
num_columns = ['маркет_актив_6_мес', 'маркет_актив_тек_мес', 'маркет_актив_тек_мес', 
               'длительность', 'средний_просмотр_категорий_за_визит',
               'неоплаченные_продукты_штук_квартал', 'ошибка_сервиса', 'страниц_за_визит', 
               'выручка_предыдущий_месяц', 'выручка_препредыдущий_месяц', 'выручка_текущий_месяц', 
               'минут_предыдущий_месяц', 'минут_текущий_месяц']
````

````python
ohe_pipe = Pipeline(
    [
        ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('ohe', OneHotEncoder(handle_unknown='error', sparse_output=False))  # Изменили sparse на sparse_output
    ]
)
````

````python
ord_pipe = Pipeline(
    [
        (
            'simple_imputer_ord_before',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',
            OrdinalEncoder(categories=[
                                      ['Редко покупает по акции','Часто покупает по акции']],
                          handle_unknown='use_encoded_value',
                          unknown_value=np.nan)
        ),
        (
            'simple_imputer_ord_after',
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
)
````

````python
data_preprocessor = ColumnTransformer(
    [('ohe', ohe_pipe, ohe_columns),
     ('ord', ord_pipe, ord_columns),
     ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)
````

````python
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=42))
])
````

````python
param_grid = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=42)],
        'models__max_depth': range(2, 7),
        'models__max_features': range(2,7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=42, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    # словарь для модели SVC()
    {
        'models': [SVC(random_state=42, kernel='poly')],
        'models__degree': range(2, 3),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
````

````python
randomized_search = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    random_state=42,
    n_jobs=-1
)
````

````python
randomized_search.fit(X_train, y_train)
````
